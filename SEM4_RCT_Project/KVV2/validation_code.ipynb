```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

edge_table = pd.read_excel(
    "/home/ibab/Downloads/Hyperphagia_Final (1).xlsx",
    sheet_name=2
)

G = nx.from_pandas_edgelist(
    edge_table,
    source="Source Node",
    target="Target Node",
    create_using=nx.DiGraph()
)
```

```python
Nodes = list(G.nodes())

Output_nodes = [node for node in Nodes if G.out_degree(node) == 0]
Input_nodes = [node for node in Nodes if G.in_degree(node) == 0]

tf = nx.is_directed_acyclic_graph(G)

if tf:
    print("No cycles detected")
else:
    print("Cycles detected")

C = list(nx.simple_cycles(G))
print(C)
```

```python
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
plt.show()
```

```python
source_node = 'Insulin'
target_node = "Hyperphagia_Phenotype"

if source_node in G.nodes() and target_node in G.nodes():
    paths = list(nx.all_simple_paths(G, source=source_node, target=target_node))
    print(paths)
else:
    print("Node not found")
```
